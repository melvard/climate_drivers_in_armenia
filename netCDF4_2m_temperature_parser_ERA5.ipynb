```python
import xarray as xr
# import matplotlib.pyplot as plt
%matplotlib inline
# import cartopy.crs as ccrs
```

```python
def get_data(file_name):
    ds = xr.open_dataset(file_name)
    return ds
```

```python
ds = get_data('./NetCDF_data/era5_minimum_2m_temperature_since_previous_post_processing_1999_11_30.nc');
```

```python
ds_c = ds.mn2t - 273
ds_c.attrs = ds.mn2t.attrs
```

```python
ds_c.attrs['units'] = 'ºC'
```

```python
def get_armenia_2m_temperature(day_hour):
    
    ds_armenia = ds_c.sel(longitude=slice(43.461914, 46.647949), latitude=slice(41.145570,38.839708))
    ds_armenia = ds_armenia.isel(time=day_hour)
    # ds_c.isel(time=0).plot(cmap='jet')
    ds_armenia.plot(cmap='jet')
    # ds_c1.plot.line(aspect=2, size=10, marker='o', color='purple')

get_armenia_2m_temperature(day_hour=16)
```

```python
ds_c2 = ds_c.isel(time=1) # 2m minimum temperature map at 1999-11-30 01:00:00 in the world 
ds_c2.plot()
```
